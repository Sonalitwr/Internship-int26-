## QUESTION 2 -Write a python program to display IMDB’s Top rated 100 hollywood movies 

```python
url2 =  "https://www.imdb.com/list/ls091520106/"
```

```python
#send get request to server
page2 = requests.get(url2)
page2
```

```python
#get content of the page 
soup2=BeautifulSoup(page2.content)
```

```python
#scrape names of the movies 
name = soup2.find_all("h3", class_="lister-item-header")
name
```

```python
# get text from movie name web elements
movies_name = [] #empty list
for i in name:
    for j in i.find_all("a"):
        movies_name.append(j.text)
movies_name
```

```python
#scrape year of release
Y_O_R=[]
for i in soup2.find_all("span",class_="lister-item-year text-muted unbold"):
    Y_O_R.append(i.text.replace('(', '').replace(')', ''))
print(Y_O_R[:100])
```

```python
#scrape rating of the movies 
R_O_M=[]
for i in soup2.find_all("div",class_="ipl-rating-star small"):
    R_O_M.append(float(i.text))

print(R_O_M)
```

```python
#framing the dataset
imdb_holly=pd.DataFrame()
imdb_holly['Name']= movies_name[:100]
imdb_holly['Year']=Y_O_R[:100]
imdb_holly['Rating']=R_O_M[:100]
imdb_holly
```

## QUESTION 4 Write a python program to scrape product name , price and discount from https://meesho.com/bags-ladies/pl/p7vbp 

```python
#send get request to server
url4='https://meesho.com/bags-ladies/pl/p7vbp'
page4=requests.get(url4)
page4
```

```python
#get content of the data 
soup4=BeautifulSoup(page4.content)
```

```python
#scrape details of the product 
name_of_product = soup4.find_all("div", class_="Card__BaseCard-sc-b3n78k-0 cXRroa NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 eEBPAI NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 eEBPAI")
```

```python
#scrape name of product from details 
nameofproduct = [] 
for i in name_of_product:
    for j in i.find_all("p"):
        nameofproduct.append(j.text)
product_name = nameofproduct[0:len(nameofproduct):3]
product_name
```

```python
#scrape discount on the product 
discount=soup4.find_all('div', class_="Card__BaseCard-sc-b3n78k-0 fVRkfg NewProductCard__PriceRow-sc-j0e7tu-5 dYYUrF NewProductCard__PriceRow-sc-j0e7tu-5 dYYUrF")
product_discount = [] 
for i in discount:
    for j in i.find_all("span"):
        product_discount.append(j.text)
product_discount
```

```python
#scrape price of the product
product_price = [] 
for i in discount:
    for j in i.find_all("h5"):
        product_price.append(j.text)
product_price 
```

```python
#framing the data set 
meesho=pd.DataFrame({})
meesho['Product_name']=product_name
meesho['Price']=product_price
meesho['Discount']=product_discount
meesho 
```
